```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
values = [0, 1, 2, 3, 4, 5, 6, 7]
squares = [value ** 2 for value in values]
plt.plot(values, squares)
plt.show()
```

```python
plt.scatter(values, squares)
plt.show()
```

```python
values = list(range(101))
pows = [value ** 3 for value in values]
plt.scatter(values, pows)
plt.show()
```

```python
plt.scatter(values, pows, c=pows, cmap=plt.cm.Blues, edgecolor='none', s=15)
plt.show()
```

```python
values = list(range(50))
squares = [pow(value, 2) for value in values]
cubs = [pow(value, 3) for value in values]

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
ax1.scatter(values, squares)
ax1.set_title('Square numbers')
ax2.plot(values, cubs)
ax2.set_title('Cube numbers')
plt.show()
```

```python
values = list(range(10))
squares = [pow(value, 2) for value in values]
cubs = [pow(value, 3) for value in values]

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.scatter(values, squares)
ax1.set_title('Square numbers')
ax2.plot(values, cubs)
ax2.set_title('Cube numbers')
plt.show()
```

```python
data = {
    'Year': [2001, 2002, 2003, 2004, 2005, 2006],
    'Sales': [100, 150, 130, 180, 210, 230]
}

df = pd.DataFrame(data)
df.plot(x='Year', y='Sales', kind='line')
plt.title('Yearly Sales')
plt.show()
```

```python
data = {
    'Category': ['First', 'First', 'Second', 'Second', 'Third', 'Third', 'Third'],
    'Values': [10, 20, 25, 45, 45, 80, 100]
}

df = pd.DataFrame(data)
df
```

```python
data = {
    'Month': ['January', 'February', 'March', 'April', 'May'],
    'Week 1 temperature': [6, 10, 12, 14, 17],
    'Week 2 temperature': [8, 12, 13, 14, 18],
    'Week 3 temperature': [9, 11, 14, 16, 20],
    'Week 4 temperature': [10, 15, 10, 17, 22]
}

df = pd.DataFrame(data)
df
```

```python
df.set_index('Month', inplace=True)
df.head()
```

```python
sns.heatmap(df, cmap='coolwarm', annot=True)
plt.title('Weekly teamerature data')
plt.show()
```
